```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
bit_df = pd.read_csv("coin_Bitcoin.csv")
```

```python
bit_df.head()
```

```python
for index, row in bit_df.iterrows():
    year = row[3]
    
    if "2013" in year:
        bit_df = bit_df.drop(index, axis=0)
```

```python
bit_df.head()
```

```python
doge_df = pd.read_csv("coin_Dogecoin.csv")
```

```python
for index, row in doge_df.iterrows():
    year = row[3]
    
    if "2013" in year:
        doge_df = doge_df.drop(index, axis=0)
```

```python
doge_df.head()
```

```python
x_ticks = []
x_labels = []

key_date = "-01-01 23:59:59"

for index, row in doge_df.iterrows():
    year = row[3]
    if key_date in year:
        x_ticks.append(index)
        x_labels.append(year[0:4])
    
fig, ax = plt.subplots(figsize=(16,8))

ax2 = ax.twinx()

mpl.style.use('seaborn')

ln = ax.plot(doge_df['Close'].values, color="red", label="Doge")
ax.set_title('Bitcoin/Dogecoin Values', fontsize=18)
ax.set_xlabel('Year', fontsize=12)
ax.set_xticks(x_ticks)
ax.set_xticklabels(x_labels, rotation=45)

ax.set_ylabel('Dogecoin Value ($)', fontsize=12, labelpad=1.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ln1 = ax2.plot(bit_df['Close'].values, label="Bitcoin")
ax2.set_ylabel('Bitcoin Value ($)', fontsize=12, labelpad=1.5)

lns = ln + ln1
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc=0)

fig.savefig("long_term.png")

```

```python
import matplotlib.pyplot as plt
from mpl_finance import candlestick_ohlc
import pandas as pd
import matplotlib.dates as mpl_dates
import numpy as np
import datetime
  
ohlc = doge_df.loc[:, ['Date', 'Open', 'High', 'Low', 'Close']]
ohlc['Date'] = pd.to_datetime(ohlc['Date'])
ohlc['Date'] = ohlc['Date'].apply(mpl_dates.date2num)
ohlc = ohlc.astype(float)
  
# Creating Subplots
fig, ax = plt.subplots()
  
candlestick_ohlc(ax, ohlc.values, width=0.6, colorup='green',
                 colordown='red', alpha=0.4)
  
# Setting labels & titles
ax.set_xlabel('Date')
ax.set_ylabel('Price')
fig.suptitle('Stock Prices of a week')
  
# Formatting Date
date_format = mpl_dates.DateFormatter('%d-%m-%Y')
ax.xaxis.set_major_formatter(date_format)
fig.autofmt_xdate()
  
fig.tight_layout()
fig.savefig("fig1.png")
plt.show()

#https://www.geeksforgeeks.org/how-to-create-a-candlestick-chart-in-matplotlib/
```

```python
def f(row):
    if row['Delta'] < 0:
        val = True 
    else:
        val = False
    return val
```

```python
bit_df['Neg'] = bit_df.apply(f, axis=1)
doge_df['Neg'] = doge_df.apply(f, axis=1)
```

```python
tt = 0
tf = 0
ft = 0
ff = 0

for index, row in bit_df.iterrows():
    for index1, row1 in doge_df.iterrows():
        if (row[3] == row1[3]):
            neg = row[11]
            neg1 = row1[11]

            if neg == True and neg1 == True:
                tt += 1
            elif neg == True and neg1 == False:
                tf += 1
            elif neg == False and neg1 == True:
                ft += 1
            elif neg == False and neg1 == False:
                ff += 1
                
hs = ([tt, tf],
      [ft, ff])

hs
```

```python
import seaborn as sns; sns.set_theme()

ax = sns.heatmap(hs)
labels = ["True", "False"]
ax.set_xticklabels(labels, rotation=45)
ax.set_yticklabels(labels, rotation=45)
ax.set_ylabel('Dogecoin Lost Value', fontsize=12, labelpad=1.5)
ax.set_xlabel('Bitcoin Lost Value', fontsize=12, labelpad=1.5)
```
